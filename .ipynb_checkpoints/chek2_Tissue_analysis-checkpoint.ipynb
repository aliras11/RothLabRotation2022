```python
chek2['rank_refined_score'] = chek2['refined_score'].rank(ascending = False) #higher scores, higher ranks
```

```python
def random_ranksum_dist(batch_size,num_samples): #batch_size is the number of ranks , #num_samples is how many samples of a certain batch_size is taken
    ranks = chek2.rank_refined_score
    rng = np.random.default_rng()
    out = [np.sum(rng.choice(ranks.to_numpy(),size=batch_size,replace=False)) for i in range(num_samples)]
    return np.array(out)
#generates a list of sums of ranks for different batch_size's this can be used to calculate p-values for actual ranks of scores for tissues
```

```python
tissues_summary = chek2.groupby('Primary_site')['Primary_site'].count()
tissues_summary_dict = tissues_summary.to_dict()
display(tissues_summary)
```

```python
ranksums_summary = chek2.groupby('Primary_site')['rank_refined_score'].sum()
ranksums_summary_dict = ranksums_summary.to_dict()
ranksums_summary
```

```python
temp = random_ranksum_dist(750,1000000)
```

```python
plt.hist(temp,bins=50)
plt.show()
```

```python
def pval_calc(tissue,num_ran_samples=1000000): #this function requires the dictionaries generated in the cells above
    temp = random_ranksum_dist(tissues_summary_dict[tissue],num_ran_samples)
    pval = sum(temp <= ranksums_summary_dict[tissue])/num_ran_samples
    return pval #returns proportion of times a rank sum less than or equal to that of a given tissue occurred
```

```python
pval_calc('salivary_gland') #TRIAL RUN
```

```python
pvals = [pval_calc(x) for x in tissues_summary_dict]
print(pvals)
```

```python
names = [x for x in tissues_summary_dict]
x = list(zip(names,pvals))
print(x)
```
